```python
import pandas as pd
import numpy as np
my_df = pd.read_csv("dog_data_short.csv")
my_df
```

```python
gender = ['Male', 'Female', 'Male', 'Male', 'Female']
my_df['Gender']=gender
my_df
```

```python
# remove column without deleting it
my_df.drop('Gender',axis = 1) # axis = 0 - rows(x), axis = 1 - columns(y) (AS X,Y fields)
```

```python
# Show DataFrame again (column is still there?)
my_df
```

```python
# Remove column with deleting it
my_df.drop('Gender', axis=1, inplace=True)
my_df
```

```python
# Remove row with deleting it (IF YOU WANT TO DELETE IT FROM DF, USE INPLACE=TRUE)
my_df.drop(0,axis=0,inplace=True) # you can just do my_df.drop(0) because axis is 0 by default
my_df
```

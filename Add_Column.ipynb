```python
import pandas as pd
import numpy as np

my_df = pd.read_csv("dog_data_short.csv")
my_df
```

```python
# Add column from List
gender = ["Male", "Female", "Male", "Male", "Female"] # same amount as rows have
my_df["Gender"] = gender
my_df
```

```python
# Add default values (if we dont know how many of them are there, almost all the time, assign the same value to all rows in that column)
my_df['Alive/Dead'] = [True]*len(my_df)
my_df
```

```python
# Add Nan (null) value !!!!!! np.nan !!!!!!
my_df['Show Dog'] = [np.nan]*len(my_df)
my_df
```

```python
# Add column using loc (not in the vid)
address = ['idk']*len(my_df)
my_df.loc[:, 'Address'] = address
my_df
```

```python
# add column using .insert - Allows position
my_df.insert(1,'Adopted', [True]*len(my_df), True) # 1st item is column's position (they start at 0), 2nd - header, 3rd - value in rows, 4th - default 
# value. will we allow duplicates or nah
my_df
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
my_df = pd.DataFrame(np.random.randn(500,4), columns=['A', 'B', 'C', 'D'])
my_df
```

```python
my_df.plot(kind='hexbin', x='A', y='B')  
```

```python
# Gridsize + other things
my_df.plot(kind='hexbin', x='A', y='B', gridsize=10, cmap='YlGn', grid=True, title='techies8')  
```

```python
# other way
my_df.plot.hexbin(x='A', y='B', gridsize=10)
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
my_df = pd.DataFrame(np.random.randn(25,4), columns=["A", 'B', 'C', 'D'])
my_df
```

```python
# Bar Plot
my_df.plot(kind='bar')
```

```python
# Bar plot with all positive values
my_df.abs().plot(kind='bar')
```

```python
# Bar Plot stacked
my_df.abs().plot(kind='bar', stacked=True)
```

```python
# Shading
my_df.abs().plot(kind='bar', stacked=True, alpha=0.5)
```

```python
# Title
my_df.abs().plot(kind='bar', stacked=True, title='techies2')
```

```python
# Legend
my_df.abs().plot(kind='bar', stacked=True, title='techies2', legend=False)
```

```python
# GRID ON BAR CHART (ew)
my_df.abs().plot(kind='bar', stacked=True, title='techies2', legend=False, grid=True)
```

```python
# Second way ??? (why using it lol?)
my_df.abs().plot.bar()
```

```python
my_df.abs().plot.bar(stacked=True, title='techies3', legend=False)
```

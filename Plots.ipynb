```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
my_df = pd.DataFrame(np.random.randn(25,4), columns=["A", "B", "C", "D"])
my_df
```

```python
# AREA PLOT
my_df.plot()
```

```python
# Area plot (2nd variation)
my_df.plot(kind='area', stacked=False)
```

```python
# Area plot using abs (module)
my_df.abs().plot(kind='area')
```

```python
# Shading 
my_df.abs().plot(kind='area', alpha=0.5) # alpha is from 0 to 1 depending on how bright or dark you want it to be, tim
```

```python
# Title
my_df.abs().plot(kind='area', alpha=0.5, title='techies')
```

```python
# Legend
my_df.abs().plot(kind='area', alpha=0.5, title='techies', legend=False) # we had legend tab before, so thats why we set it to false and it disappears
```

```python
# Plot only one column ["A"]
my_df['A'].abs().plot(kind='area', alpha=0.5, title='techies', legend=False)
```

```python
# Plot two columns
my_df[['A', 'B']].abs().plot(kind='area', alpha=0.5, title='techies', legend=False)
```

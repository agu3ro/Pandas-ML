```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
my_df = pd.DataFrame(np.random.randn(25,4), columns=['A','B','C','D'])
my_df
```

```python
my_df.abs().plot(kind='line')
```

```python
# Width
my_df.abs().plot(kind='line', lw=3)
```

```python
# Size
my_df.abs().plot(kind='line', lw=3, figsize=(10,5)) # figsize == tuple()
```

```python
my_df.abs().plot(kind='line', lw=3, figsize=(10,5), title='techies4', legend=False, alpha=0.6)
```

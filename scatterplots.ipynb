```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
my_df = pd.DataFrame(np.random.randn(500,4), columns=['A', 'B', 'C', 'D'])
my_df
```

```python
# Scatter plot
my_df.plot(kind='scatter', x='A', y='B', c='C')
```

```python
# C MAP (colors https://matplotlib.org/stable/users/explain/colors/colormaps.html) + some stuff from other vid
my_df.plot(kind='scatter', x='A', y='B', c='C', cmap='copper', alpha=0.6, figsize=(10,5))
```

```python
# SIZE instead of figsize (s)
my_df.plot(kind='scatter', x='A', y='B', s=my_df['C']*25, alpha=0.6)
```
